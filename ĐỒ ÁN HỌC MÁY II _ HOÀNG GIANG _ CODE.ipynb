```python
import numpy as np
from sklearn.cluster import KMeans

def initialize_gmm(data, k=3, random_state=42):

    #  Khá»Ÿi táº¡o trung tÃ¢m cá»¥m (Î¼) báº±ng K-means++
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=random_state)
    kmeans.fit(data)
    mu = kmeans.cluster_centers_  # Shape: [k, n_features]

    # Get the number of features from the data shape
    n_features = data.shape[1]

    #  Khá»Ÿi táº¡o ma tráº­n hiá»‡p phÆ°Æ¡ng sai (Î£) lÃ  ma tráº­n Ä‘Æ¡n vá»‹
    sigma = [np.eye(n_features) for _ in range(k)]  # Danh sÃ¡ch k ma tráº­n [n_features, n_features]

    #  Khá»Ÿi táº¡o trá»ng sá»‘ cá»¥m (Ï€) báº±ng phÃ¢n phá»‘i Ä‘á»u
    pi = np.ones(k) / k  # Shape: [k]

    return mu, sigma, pi

data_cpa = data

# Gá»i hÃ m khá»Ÿi táº¡o
mu, sigma, pi = initialize_gmm(data, k=3)

mu
```

```python
from scipy.stats import multivariate_normal

def expectation_step(data, mu, sigma, pi):

    n_samples, n_features = data.shape
    k = len(pi)

    # Khá»Ÿi táº¡o máº£ng gamma Ä‘á»ƒ lÆ°u xÃ¡c suáº¥t thuá»™c cá»¥m
    gamma = np.zeros((n_samples, k))

    # TÃ­nh máº­t Ä‘á»™ xÃ¡c suáº¥t cá»§a má»—i Ä‘iá»ƒm dá»¯ liá»‡u Ä‘á»‘i vá»›i má»—i cá»¥m
    for i in range(k):
        pdf = multivariate_normal.pdf(data, mean=mu[i], cov=sigma[i])  # HÃ m máº­t Ä‘á»™ xÃ¡c suáº¥t Gaussian
        gamma[:, i] = pi[i] * pdf  # Trá»ng sá»‘ * Máº­t Ä‘á»™ xÃ¡c suáº¥t

    # Chuáº©n hÃ³a Ä‘á»ƒ cÃ³ tá»•ng xÃ¡c suáº¥t báº±ng 1
    gamma_sum = np.sum(gamma, axis=1)[:, np.newaxis]  # Tá»•ng xÃ¡c suáº¥t theo hÃ ng
    gamma /= gamma_sum  # Chia cho tá»•ng xÃ¡c suáº¥t Ä‘á»ƒ chuáº©n hÃ³a

    return gamma

gamma = expectation_step(data, mu, sigma, pi)

gamma
```

```python
def maximization_step(data, gamma):

    n_samples, n_features = data.shape
    k = 3

    # Cáº­p nháº­t trung tÃ¢m cá»¥m (mu)
    mu = np.dot(gamma.T, data) / np.sum(gamma, axis=0)[:, np.newaxis]

    # Cáº­p nháº­t ma tráº­n hiá»‡p phÆ°Æ¡ng sai (sigma)
    sigma = []
    for i in range(k):
        diff = data - mu[i]
        sigma_k = np.dot(gamma[:, i] * diff.T, diff) / np.sum(gamma[:, i])
        sigma.append(sigma_k)

    # Cáº­p nháº­t trá»ng sá»‘ cá»¥m (pi)
    pi = np.sum(gamma, axis=0) / n_samples

    return mu, sigma, pi

mu_new, sigma_new, pi_new = maximization_step(data, gamma)
mu_new
```

```python
def log_likelihood(data, mu, sigma, pi):

    n_samples = data.shape[0]
    k = len(pi)
    likelihood = np.zeros((n_samples, k))

    # TÃ­nh máº­t Ä‘á»™ xÃ¡c suáº¥t cá»§a má»—i Ä‘iá»ƒm dá»¯ liá»‡u Ä‘á»‘i vá»›i má»—i cá»¥m
    for i in range(k):
        likelihood[:, i] = pi[i] * multivariate_normal.pdf(data, mean=mu[i], cov=sigma[i])

    # TÃ­nh log-likelihood tá»•ng há»£p
    log_likelihood = np.sum(np.log(np.sum(likelihood, axis=1)))

    return log_likelihood

def check_convergence(log_likelihood_old, log_likelihood_new, tol=1e-4, max_iter=100, iter_count=1):

    if iter_count >= max_iter:
        print("ÄÃ£ Ä‘áº¡t sá»‘ vÃ²ng láº·p tá»‘i Ä‘a.")
        return True

    # Kiá»ƒm tra sá»± thay Ä‘á»•i log-likelihood
    if np.abs(log_likelihood_new - log_likelihood_old) < tol:
        print("Thuáº­t toÃ¡n Ä‘Ã£ há»™i tá»¥.")
        return True

    return False

# VÃ­ dá»¥ sá»­ dá»¥ng
max_iter = 100
tol = 1e-4
iter_count = 0
log_likelihood_old = -np.inf

# Giáº£ sá»­ chÃºng ta thá»±c hiá»‡n vÃ²ng láº·p EM
while True:
    iter_count += 1

    # BÆ°á»›c E: TÃ­nh xÃ¡c suáº¥t thuá»™c cá»¥m (gamma)
    gamma = expectation_step(data_cpa, mu, sigma, pi)

    # BÆ°á»›c M: Cáº­p nháº­t tham sá»‘ mÃ´ hÃ¬nh
    mu, sigma, pi = maximization_step(data_cpa, gamma)

    # TÃ­nh log-likelihood
    log_likelihood_new = log_likelihood(data_cpa, mu, sigma, pi)

    # Kiá»ƒm tra Ä‘iá»u kiá»‡n dá»«ng
    if check_convergence(log_likelihood_old, log_likelihood_new, tol, max_iter, iter_count):
        break

    log_likelihood_old = log_likelihood_new

```

```python
def aic_bic(data, mu, sigma, pi):

    n_samples = data.shape[0]
    k = len(pi)

    # TÃ­nh log-likelihood
    log_likelihood_value = log_likelihood(data, mu, sigma, pi)

    # Sá»‘ tham sá»‘ mÃ´ hÃ¬nh
    n_params = k * (data.shape[1] * (data.shape[1] + 1) / 2) + k * data.shape[1] + k - 1

    # AIC vÃ  BIC
    aic = 2 * n_params - 2 * log_likelihood_value
    bic = np.log(n_samples) * n_params - 2 * log_likelihood_value

    return aic, bic

# TÃ­nh AIC vÃ  BIC
aic_value, bic_value = aic_bic(data_cpa, mu, sigma, pi)
print(f"AIC: {aic_value}, BIC: {bic_value}")
```

```python
def plot_bic_aic(k_range, bic_scores, aic_scores):
    plt.figure(figsize=(8, 5))
    plt.plot(k_range, bic_scores, marker='o', label='BIC', color='blue')
    plt.plot(k_range, aic_scores, marker='s', label='AIC', color='red')
    plt.xlabel('Sá»‘ lÆ°á»£ng thÃ nh pháº§n Gaussian (K)')
    plt.ylabel('GiÃ¡ trá»‹')
    plt.title('Biá»ƒu Ä‘á»“ Ä‘Ã¡nh giÃ¡ BIC vÃ  AIC')
    plt.legend()
    plt.grid()
    plt.show()

plot_bic_aic(k_range, bic_scores, aic_scores)
```
